```python
psa_pop_df = psa_pop_df_og[["YearIssued", "SetName", "SubjectName", "Variety", "GradeTotal", "Grade10", "Grade9", "set_id"]]
psa_pop_df.head()
```

```python
import pandas as pd

grouped_df = psa_pop_df.groupby('set_id').agg({'GradeTotal': 'sum', 'SetName': 'first', 'YearIssued': 'first' }).reset_index()
grouped_df.head()
```

```python
sorted_grouped_df = grouped_df.sort_values('GradeTotal', ascending=True)
sorted_grouped_df = sorted_grouped_df[sorted_grouped_df["GradeTotal"] >= 20]
sorted_grouped_df.head(20)
```
